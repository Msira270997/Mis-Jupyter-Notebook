## One hot encoding
Usaremos el dataset de Recursos Humanos

```python
dfrh = pd.read_csv("../Jupyter Notebook/Datasets/Recursos_Humanos/HRDataset_v14.csv")
dfrh.head()
```

```python
# Hacemos una seleccion de columnas para hacer un analisis un poco mas enfocado

dfrh1 = dfrh[["Employee_Name","Salary","Position","State","Sex","DOB","MaritalDesc","CitizenDesc",
              "Department","PerformanceScore","EngagementSurvey","EmpSatisfaction"]]
dfrh1.head()
```

```python
dfrh1.shape
```

```python
dfrh1.describe()
```

```python
# Usamos el one hot encoding con la columna de sexo, que crear dos columnas: masculino y femenino
# Creamos un objeto 

dummy_sex = pd.get_dummies(dfrh1["Sex"], prefix="sexo")
```

```python
dummy_sex.head()
```

```python
# Vamos a eliminar la columna de sex

dfrh1 = dfrh1.drop(["Sex"], axis = 1)
```

```python
dfrh1.shape
```

```python
dfrh1.head(3)
```

```python
# Ahora vamos a agregar las columnas de sex (dummy_sex)
# Creamos un nuevo dataset y le concatenamos el anterior con las columnas nuevas

dfrh2 = pd.concat([dfrh1, dummy_sex], axis = 1)
```

```python
dfrh2.shape
```

```python
dfrh2.head(4)
```

```python
# Ahora vamos a automatizar el mismo proceso con una funcion tomando como ejemplo CitizenDesc

def createdummies(df, varname): # Creamos una funcion, los parametros son el dataframe y la columna a aplicar one hot encoding
    dummy=pd.get_dummies(df[varname], prefix = varname) # Convertimos la variable categorica a variable numerica o contable
    df=df.drop([varname], axis = 1) # Eliminamos la columna de la variable categorica
    df=pd.concat([df, dummy], axis = 1) # Concatenamos el dataframe con las culumnas de la variable contable
    return df.head(4)
```

```python
createdummies(dfrh2, "CitizenDesc")
```

### Funcion Merge (unir)

```python
# Vamos a crear un dataframe (Primera forma)

staff_df = pd.DataFrame({'Name':['Kelly','Sally','James'],
                         'Role':['Director of RH','Course Liasion','Grader']})
staff_df
```

```python
# La funcion .set_index() se utiliza para establecer columnas como indice del dataframe

staff_df = staff_df.set_index('Name')
staff_df
```

```python
# Creamos el segundo dataframe (Segunda forma)

student_df = pd.DataFrame([{'Name':'James','School':'Business'},
                          {'Name':'Myke','School':'Law'},
                          {'Name':'Sally','School':'Engineering'}])
student_df
```

```python
student_df = student_df.set_index('Name')
student_df
```

```python
print(staff_df)
print(student_df)
```

### Outer
Aqui obtenemos un dataset con todos los registros de los dos datasets

```python
pd.merge(staff_df, student_df, how = 'outer', left_index = True, right_index = True)
```

### Inner
Aqui obtenemos un dataset con los registros en comun de los dos datasets

```python
pd.merge(staff_df, student_df, how = 'inner', left_index = True, right_index = True)
```

### Left
Aqui obtenemos un dataset con los registros del dataset izquierdo y con la informacion de las columnas correspondientes al dataset derecho

```python
pd.merge(staff_df, student_df, how = 'left', left_index = True, right_index = True)
```

### Right
Aqui obtenemos un dataset con los registros del dataset derecho y con la informacion de las columnas correspondientes al dataset izquierdo

```python
pd.merge(staff_df, student_df, how = 'right', left_index = True, right_index = True)
```

### Right sin index

```python
# Aqui vamos a eliminar los indices asignados previamente en los dos datasets

staff_df = staff_df.reset_index()
student_df = student_df.reset_index()
```

```python
# Aqui obtenemos un dataset con los registros del dataset derecho y con la informacion de las columnas correspondientes
# al dataser izquierdo

pd.merge(staff_df, student_df, how = 'right', on = 'Name')
```
