### Ahora vamos a obtener los followers de un unsuario
En este caso de la cuenta nike

```python
followers_nike = api.followers(screen_name="nike")
for user in followers_nike:
    print(json.dumps(user._json, indent=2))
  
print("Longitud del objetos de followers")
print(len(followers_nike))
```

**Como vimos anteriormente la cuenta de nike tiene 9.401.159 follwers, pero al querer obtenerlos solo me devuelve 20**

**Para obtener la cantidad de usuarios que se desee se usa el metodo cursor de la siguiente manera:**

1.   tweepy.Cursor()
2.   Los parametros dentro de corsos seran el metodo que se quiere buscar en este caso objeto.followers
3.   Los parametros del metodo que estoy llamando en este caso screem_name.
4.   Luego cuantos items (en este caso followers) quiero obtener
5.   Esto devuelve un iterable, es decir, que para obtener los datos tengo que recorrerlo en un bucle for



```python
for user in tweepy.Cursor(api.followers, screen_name="nike").items(100):
    print(json.dumps(user._json, indent=2))
```

### Ahora vamos a obtener las personas a las que nike sigue, es decir, los friends/followees

```python
for user in tweepy.Cursor(api.friends, screen_name="nike").items(2):
    print(json.dumps(user._json, indent=2))
```

# Ahora en vez de usuarios vamos a analizar tweets o timelines
Usaremos en Cursor el parametro tweet_mode que determina la longitud del tweet, en este caso usamos extended para obtener todo el tweet

```python
for tweet in tweepy.Cursor(api.user_timeline, screen_name="neymarjr", tweet_mode="extended").items(1):
    print(json.dumps(tweet._json, indent=2))
```

### Ahora vamos a buscar tweets con el metodo search

```python
for tweet in tweepy.Cursor(api.search, q="venezuela odoo", tweet_mode="extended").items(10):
    print(json.dumps(tweet._json, indent=2))
```

```python
for status in tweepy.Cursor(api.search, "petro", count=100).items(10):
        print(json.dumps(status.text, indent=2))
```

**De esta forma guardamos los datos en formato json en un archivo de texto o en este caso con extencion .json**

```python
neymar_user = api.get_user("neymarjr")

with open('data.json', 'w') as text:
    json.dump(neymar_user._json, text, indent=2)
```
